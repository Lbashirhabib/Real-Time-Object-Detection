```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from PIL import Image
import os

print(" All packages installed successfully!")
print(f"PyTorch version: {torch.__version__}")
print(f"OpenCV version: {cv2.__version__}")
```

Now we are to load our pre-trained YOLO model

```python
# Load different YOLO model sizes - choosing one based on our needs
model = YOLO('yolov8n.pt')  # Nano (fastest, lowest accuracy)
# model = YOLO('yolov8s.pt')  # Small
# model = YOLO('yolov8m.pt')  # Medium  
# model = YOLO('yolov8l.pt')  # Large
# model = YOLO('yolov8x.pt')  # XLarge (slowest, highest accuracy)

print(" YOLO model loaded successfully!")
print(f"Model type: {model.__class__.__name__}")

# Check if GPU is available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Move model to GPU if available
model.to(device)
```

```python
# Let Download a sample image for testing
import requests

def download_sample_image():
    """Download a sample image for object detection testing"""
    url = "https://raw.githubusercontent.com/ultralytics/ultralytics/main/ultralytics/assets/bus.jpg"
    response = requests.get(url)
    
    with open('sample_image.jpg', 'wb') as f:
        f.write(response.content)
    
    print(" Sample image downloaded successfully!")
    return 'sample_image.jpg'

# Download the image
image_path = download_sample_image()

# Display the original image
original_image = Image.open(image_path)
plt.figure(figsize=(10, 6))
plt.imshow(original_image)
plt.title("Original Image - Before Detection")
plt.axis('off')
plt.show()
```

```python
# Perform object detection on the image
print(" Performing object detection...")

# Run inference on the image
results = model(image_path)

# results contains all detection information
print(f"Number of detections: {len(results)}")

# Get the first result (since we only processed one image)
result = results[0]

# Print detection details
print(f"Image shape: {result.orig_shape}")
print(f"Number of objects detected: {len(result.boxes)}")

# Show what classes were detected
if len(result.boxes) > 0:
    print("\n Detected Objects:")
    for i, box in enumerate(result.boxes):
        class_id = int(box.cls[0])
        class_name = model.names[class_id]
        confidence = float(box.conf[0])
        print(f"  {i+1}. {class_name}: {confidence:.2%} confident")
```

```python
# Lets Visualize Result
# Method 1: Using YOLO's built-in plotting (easiest)
print(" Visualizing results...")

# Plot results on image
annotated_image = result.plot()  # This adds bounding boxes and labels

# Convert BGR to RGB for matplotlib
annotated_image_rgb = annotated_image[..., ::-1]

# Display the annotated image
plt.figure(figsize=(12, 8))
plt.imshow(annotated_image_rgb)
plt.title("YOLOv8 Object Detection Results")
plt.axis('off')
plt.show()

# Save the annotated image
cv2.imwrite('detected_objects.jpg', annotated_image)
print(" Annotated image saved as 'detected_objects.jpg'")
```

```python
# Let's examine the detection data in more detail
print("\n Detailed Detection Analysis:")
print("=" * 50)

for i, box in enumerate(result.boxes):
    # Get box coordinates (x1, y1, x2, y2)
    coordinates = box.xyxy[0].cpu().numpy()
    x1, y1, x2, y2 = coordinates.astype(int)
    
    # Get class information
    class_id = int(box.cls[0])
    class_name = model.names[class_id]
    confidence = float(box.conf[0])
    
    # Calculate box dimensions
    width = x2 - x1
    height = y2 - y1
    
    print(f"Object {i+1}:")
    print(f"  Class: {class_name} (ID: {class_id})")
    print(f"  Confidence: {confidence:.2%}")
    print(f"  Bounding Box: ({x1}, {y1}) to ({x2}, {y2})")
    print(f"  Dimensions: {width} x {height} pixels")
    print(f"  Area: {width * height} pixels")
    print("-" * 30)
```

Real-Time webcam detection

```python
def real_time_detection():
    """Perform real-time object detection using webcam"""
    print(" Starting real-time detection...")
    print("Press 'q' to quit, 's' to save current frame")
    
    # Initialize webcam
    cap = cv2.VideoCapture(0)
    
    if not cap.isOpened():
        print(" Error: Could not open webcam")
        return
    
    # Set camera resolution
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    
    frame_count = 0
    saved_count = 0
    
    while True:
        # Read frame from webcam
        ret, frame = cap.read()
        if not ret:
            print(" Error: Could not read frame")
            break
        
        # Perform object detection on the frame
        results = model(frame)
        
        # Annotate the frame with detections
        annotated_frame = results[0].plot()
        
        # Add FPS counter
        fps = cap.get(cv2.CAP_PROP_FPS)
        cv2.putText(annotated_frame, f'FPS: {fps:.1f}', (10, 30), 
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        # Display the annotated frame
        cv2.imshow('YOLOv8 Real-Time Detection', annotated_frame)
        
        # Handle key presses
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):  # Quit
            break
        elif key == ord('s'):  # Save current frame
            cv2.imwrite(f'saved_frame_{saved_count}.jpg', annotated_frame)
            print(f" Frame saved as 'saved_frame_{saved_count}.jpg'")
            saved_count += 1
        
        frame_count += 1
    
    # Clean up
    cap.release()
    cv2.destroyAllWindows()
    print(f" Real-time detection completed. Processed {frame_count} frames.")

# Uncomment the line below to run real-time detection
# real_time_detection()
```

Process Video Files

```python
def process_video(video_path, output_path='output_video.mp4'):
    """Process a video file with object detection"""
    print(f" Processing video: {video_path}")
    
    # Open video file
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print(f" Error: Could not open video {video_path}")
        return
    
    # Get video properties
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    print(f"Video info: {width}x{height}, {fps} FPS, {total_frames} frames")
    
    # Setup video writer for output
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
    
    frame_count = 0
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Perform detection
        results = model(frame)
        annotated_frame = results[0].plot()
        
        # Write frame to output video
        out.write(annotated_frame)
        
        frame_count += 1
        if frame_count % 30 == 0:  # Print progress every 30 frames
            progress = (frame_count / total_frames) * 100
            print(f"Progress: {progress:.1f}% ({frame_count}/{total_frames})")
    
    # Clean up
    cap.release()
    out.release()
    print(f" Video processing complete! Output saved as: {output_path}")
    print(f"Processed {frame_count} frames")

# Example usage (you'll need a video file):
# process_video('input_video.mp4', 'detected_video.mp4')
```

Custom object detection

```python
def detect_objects(image_path, confidence_threshold=0.5, save_result=True):
    """
    Custom object detection function with configurable confidence threshold
    
    Args:
        image_path: Path to input image
        confidence_threshold: Only show detections above this confidence (0-1)
        save_result: Whether to save the annotated image
    """
    # Load and process image
    image = cv2.imread(image_path)
    results = model(image, conf=confidence_threshold)  # Apply confidence threshold
    
    # Get detections
    result = results[0]
    annotated_image = result.plot()
    
    # Display
    plt.figure(figsize=(12, 8))
    plt.imshow(annotated_image[..., ::-1])
    plt.title(f"Object Detection (Confidence  {confidence_threshold})")
    plt.axis('off')
    plt.show()
    
    # Save if requested
    if save_result:
        output_path = f"detection_conf_{confidence_threshold}.jpg"
        cv2.imwrite(output_path, annotated_image)
        print(f" Saved as: {output_path}")
    
    # Print summary
    print(f"\n Detection Summary:")
    print(f"Confidence threshold: {confidence_threshold}")
    print(f"Objects detected: {len(result.boxes)}")
    
    return result

# Test with different confidence levels
print("Testing with high confidence (0.7):")
high_conf_result = detect_objects(image_path, confidence_threshold=0.7)

print("\nTesting with low confidence (0.3):")
low_conf_result = detect_objects(image_path, confidence_threshold=0.3)
```

```python
# COMPLETE OBJECT DETECTION SCRIPT
def complete_object_detection_demo():
    """Complete demo of YOLOv8 object detection capabilities"""
    
    print(" YOLOv8 Object Detection Demo")
    print("=" * 40)
    
    # 1. Setup
    model = YOLO('yolov8n.pt')
    
    # 2. Download sample image
    image_path = download_sample_image()
    
    # 3. Perform detection
    results = model(image_path)
    result = results[0]
    
    # 4. Display results
    annotated_image = result.plot()
    plt.figure(figsize=(15, 10))
    plt.imshow(annotated_image[..., ::-1])
    plt.title("YOLOv8 Object Detection - Complete Demo")
    plt.axis('off')
    plt.show()
    
    # 5. Print detailed information
    print(f"\n Detection Results:")
    print(f"Total objects detected: {len(result.boxes)}")
    
    for i, box in enumerate(result.boxes):
        class_id = int(box.cls[0])
        class_name = model.names[class_id]
        confidence = float(box.conf[0])
        print(f"  {i+1}. {class_name}: {confidence:.2%}")
    
    print(f"\n Try running real_time_detection() for live webcam detection!")
    print(" Or process_video() for video file processing!")

# Run the complete demo
complete_object_detection_demo()
```
